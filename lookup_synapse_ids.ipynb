```python
import pandas as pd
```

```python
# this csv file is 379 MB and too large to place on Github
# in order to run this notebook, you must first download the file and place on your local machine in the data/ folder
# the csv file is available here: https://www.microns-explorer.org/phase1 (scroll down to synapse table section)

synapses = pd.read_csv("pni_synapses_v185.csv")
```

```python
synapses
```

```python
# query: what cell has the most pre-synaptic sites?
# First, do a value count on pre_root_id
synapses["pre_root_id"].value_counts().max()
```

```python
# Next, find the cellid for this cell
mostpre = synapses["pre_root_id"].value_counts()
```

```python
mostpre.to_frame("pre_count").head(10)
```

```python
# query: what cell has the most post-synaptic sites?
# First, do a value count on post_root_id
synapses["post_root_id"].value_counts().max()
```

```python
mostpost = synapses["post_root_id"].value_counts()
```

```python
mostpost.to_frame("post_count").head(10)
```

```python
# define a variable for this cellid

cellidlookup0 = 648518346349538440
```

This is the excitatory pyramidal neuron with the most post-synaptic sites in the volume;  
Identified using the cellid lookup table, for example use this notebook to look up cell subtypes:  
https://github.com/shandran/layer23-volume/blob/main/lookup_cellid_in_layer23_volume.ipynb

```python
query0pre = synapses[synapses["pre_root_id"] == cellidlookup0]
query0post = synapses[synapses["post_root_id"] == cellidlookup0]
```

```python
len(query0pre)
```

This excitatory pyramidal neuron has 106 synapses located in its axon (hence these are pre-synaptic sites);  
In other words, this neuron is making 106 synapses onto other processes in the volume;  
(these are all making synaptic contacts with processes that are contained within the volume)

```python
len(query0post)
```

This excitatory pyramidal neuron has 5369 synapses in its dendrites and soma;  
In other words, this neuron is receiving 5369 synapses from other processes;  
(these could be from neurons located inside or outside the volume)
