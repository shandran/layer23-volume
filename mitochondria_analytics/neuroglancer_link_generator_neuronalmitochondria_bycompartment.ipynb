```python
query = mito[mito["cellid"] == lookup]
len(query)
```

```python
query
```

```python
# enter compartment of interest
# choices: 'Somatic', 'Axonal', 'Apical', 'Basal', 'Unknown dendritic', 'Unknown'

comp = 'Axonal'
```

```python
query_comp = query[query['compartment'] == comp]
query_comp
```

```python
query_list = query_comp["mito_id"].to_list()
# query_list
```

```python
print(query_list)
```

```python
spacer = "%22%2C%22"
query_list_join = spacer.join(map(str,query_list))
```
