```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None)
```

```python
# import rfc results file

rfc_results = pd.read_csv('unknown_dendritic_rfc_prediction_results_withincone_replicate_version.csv', index_col=[0])
rfc_results
```

```python
# import human curated dataframe to check the predictions against

human_curated = pd.read_csv('human_curated_compare.csv', index_col=[0])
human_curated
```

```python
# reduce to mito_id and human_assignment columns for merge with rfc prediction df

human_curated_merge = human_curated[['mito_id','human_assignment']]
human_curated_merge
```

```python
# merge dfs

rfc_results_check = pd.merge(rfc_results, human_curated_merge, on='mito_id', how='inner')
rfc_results_check
```

```python
rfc_results_check['correct'] = np.where(rfc_results_check['rfc_predicted_compartment'] == rfc_results_check['human_assignment'], 1, 0)
rfc_results_check
```

```python
accuracy = sum(rfc_results_check.correct) / len(rfc_results_check)
print(np.round(accuracy, 4))
```

```python
# compare to within45cone results

rfc_results_withincone = pd.read_csv('unknown_dendritic_rfc_prediction_results_withincone.csv', index_col=[0])
rfc_results_withincone
```

```python
# merge df for withincone

rfc_results_withincone_check = pd.merge(rfc_results, human_curated_merge, on='mito_id', how='inner')
rfc_results_withincone_check
```

```python
rfc_results_withincone_check['correct'] = np.where(rfc_results_check['rfc_predicted_compartment'] == rfc_results_check['human_assignment'], 1, 0)
rfc_results_withincone_check
```

```python
accuracy = sum(rfc_results_check.correct) / len(rfc_results_check)
print(np.round(accuracy, 4))
```
