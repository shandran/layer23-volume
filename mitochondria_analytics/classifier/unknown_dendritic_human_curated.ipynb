```python
# import pyr_all.csv containing mito analysis file with Turner et al Cell 2022 compartment designations
# this is for carrying over the original assignment label when updating future dfs for additional cycles of learning

pyr_all = pd.read_csv('mito_above_soma_within45cone.csv', index_col=[0])
pyr_all
```

```python
# isolate mito_id and compartment columns for merge with human curated df

pyr_all_orig_compartment = pyr_all[['mito_id', 'compartment']]

# rename compartment column to 'Turner_compartment_assignment'

pyr_all_orig_compartment.rename(columns = {'compartment':'Turner_compartment_assignment'}, inplace = True)
pyr_all_orig_compartment
```

```python
human_curated_compare = pd.merge(human_curated, pyr_all_orig_compartment, on='mito_id', how='inner')
human_curated_compare
```

```python
human_curated_compare.human_assignment.value_counts()
```

```python
# isolate somatic assignments
# these will need to be reassigned in an updated training dataset

somatic_assignments = human_curated_compare[human_curated_compare['human_assignment'] == "Somatic"]
somatic_assignments
```

```python
len(somatic_assignments)
```

```python
# isolate the new "True unknown dendritic" assignments
# These are true unknown dendritic that can't be determined by human inspection let alone a classifier

true_unkdend_assignments = human_curated_compare[human_curated_compare['human_assignment'] == "True unknown dendritic"]
true_unkdend_assignments
```

```python
# drop somatic and true_unknown_dendritic

apical_basal = human_curated_compare[(human_curated_compare['human_assignment'] == "Apical") | (human_curated_compare['human_assignment'] == "Basal")]
apical_basal
```

```python
print(f"classifier success with confounders removed: {(np.round(apical_basal.rfc_correct.sum() / len(apical_basal), 4))}")
```

# Create an updated training df with the human curated assignments

```python
# drop all columns except mito_id and human_assignment for merging with pyr_all

human_curated_assignments = human_curated_compare[['mito_id','human_assignment']]
human_curated_assignments
```

```python
human_curated_assignments.human_assignment.value_counts()
```

```python
mito_above_soma_withincone_human_assignment = pd.merge(pyr_all, human_curated_assignments, on='mito_id', how='outer')
mito_above_soma_withincone_human_assignment
```

```python
mito_above_soma_withincone_human_assignment.human_assignment.value_counts(dropna=False)
```

```python
# add a new column 'updated_compartment' transferring the human-curated assignments, and if NaN, copy the original compartment assignment

mito_above_soma_withincone_human_assignment['updated_compartment'] = np.where(
    mito_above_soma_withincone_human_assignment['human_assignment'].isnull(), 
    mito_above_soma_withincone_human_assignment['compartment'], 
    mito_above_soma_withincone_human_assignment['human_assignment']  
)

mito_above_soma_withincone_human_assignment
```

```python
mito_above_soma_withincone_human_assignment.updated_compartment.value_counts()
```
