# List of all reciprocal pairs in the Layer 2/3 volume
Reciprocal pairs means a list of cell ids where one cell id synapses onto another one AND vice versa  
In other words, there are one or more pre- to post-synaptic sites from cell id 1 to cell id 2  
AND ALSO, there are one or more pre- to post-synaptic sites from cell id 2 to cell id 1

I am grateful to Wanwen Zeng at Stanford University for writing the code to create a dictionary to search for reciprocal pairs that generated the `paired.csv` dataframe below.

```python
import pandas as pd
import numpy as np
```

```python
colnames=["id", "pre_root_id", "post_root_id", "cleft_vx", "ctr_pt_x_nm", "ctr_pt_y_nm", "ctr_pt_z_nm", "pre_pos_x_vx", "pre_pos_y_vx", "pre_pos_z_vx", "ctr_pos_x_vx", "ctr_pos_y_vx", "ctr_pos_z_vx", "post_pos_x_vx", "post_pos_y_vx", "post_pos_z_vx"] 
pairs = pd.read_csv("paired.csv",  names=colnames, header=None)
```
