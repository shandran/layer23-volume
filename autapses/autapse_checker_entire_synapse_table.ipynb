# Check for autapses in the Layer 2/3 synapse table

The full synapse table (pni_synapses_v185.csv) is too large to place on github

```python
import pandas as pd
```

```python
synapses = pd.read_csv("data/pni_synapses_v185.csv")
```

```python
synapses
```

```python
synapses["preroot-postroot"] = synapses["pre_root_id"] - synapses["post_root_id"]
```

```python
synapses
```

```python
autapses = synapses["preroot-postroot"] == 0
```

```python
autapses.value_counts()
```

```python
autapse_df = synapses.loc[autapses]
autapse_df.sort_values(by="cleft_vx", ascending=False)
```
